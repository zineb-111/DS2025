```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
wine_quality = fetch_ucirepo(id=186)

# data (as pandas dataframes)
X = wine_quality.data.features
y = wine_quality.data.targets

# metadata
print(wine_quality.metadata)

# variable information
print(wine_quality.variables)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combine features (X) and target (y) into a single DataFrame for easier plotting
df_combined = pd.concat([X, y], axis=1)

# Plotting the distribution of 'quality'
plt.figure(figsize=(8, 6))
sns.histplot(df_combined['quality'], bins=range(0, 11), kde=True)
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality Score')
plt.ylabel('Number of Samples')
plt.xticks(range(0, 11))
plt.show()
```

Next, let's visualize the correlation matrix to see how all the variables relate to each other, especially to the `quality` variable.

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df_combined.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Wine Quality Dataset')
plt.show()
```
